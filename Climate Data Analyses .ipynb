```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

#making dataframe
df = pd.read_csv('climate data.csv')

df.to_csv('irradiancedata.csv',columns=['hour','irradiance'])
New_data_solar = pd.read_csv('irradiancedata.csv')

df.to_csv('wind.csv',columns=['hour','wind speed'])
New_data_wind = pd.read_csv('wind.csv')

df.to_csv('tempdata.csv',columns=['hour','temp'])
New_data_temp = pd.read_csv('tempdata.csv')

#clustering for irradiance/hour
New_data_solar = New_data_solar.drop(columns= New_data_solar.columns[0])

kmeanssoalar = KMeans(n_clusters=5).fit(New_data_solar)
centroidssolar = kmeanssoalar.cluster_centers_
print(centroidssolar)

#clustering for irradiance/hour
New_data_wind = New_data_wind.drop(columns= New_data_wind.columns[0])
kmeanswind = KMeans(n_clusters=5).fit(New_data_wind)
centroidswind = kmeanswind.cluster_centers_
print(centroidswind)

#clustring for tempreture/hour
New_data_temp = New_data_temp.drop(columns= New_data_temp.columns[0])
kmeanstemp = KMeans(n_clusters=10).fit(New_data_temp)
centroidtemp = kmeanstemp.cluster_centers_
print(centroidtemp)

####plot
plt.scatter(New_data_solar['hour'], New_data_solar['irradiance'], c= kmeanssoalar.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidssolar[:, 0], centroidssolar[:, 1], c='red', s=50)
plt.show()

plt.scatter(New_data_wind['hour'], New_data_wind['wind speed'], c= kmeanswind.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidswind[:, 0], centroidswind[:, 1], c='red', s=50)
plt.show()


plt.scatter(New_data_temp['hour'], New_data_temp['temp'], c= kmeanstemp.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroidtemp[:, 0], centroidtemp[:, 1], c='red', s=50)
plt.show()
```
